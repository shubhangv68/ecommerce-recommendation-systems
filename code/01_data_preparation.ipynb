```python
cols = ["category", "no_of_traning_records", "no_of_valid_records","no_of_test_records", \
        "user_overlap%_valid", "product_overlap%_valid", "user_and_product_overlap%_valid",\
        "user_overlap%_test", "product_overlap%_test", "user_and_product_overlap%_test"]

rows_list = []
for category in processed_categories:
    train_df = pd.read_parquet(f'drive/MyDrive/ShoppingPulse/datasets/processed/interactions_train/train_{category}.parquet')
    valid_df = pd.read_parquet(f'drive/MyDrive/ShoppingPulse/datasets/processed/interactions_valid/valid_{category}.parquet')
    test_df = pd.read_parquet(f'drive/MyDrive/ShoppingPulse/datasets/processed/interactions_test/test_{category}.parquet')

    no_of_traning_records = len(train_df)
    no_of_valid_records = len(valid_df)
    no_of_test_records = len(test_df)
    user_overlap_valid = valid_df.user_in_train.sum() * 100 // len(valid_df)
    product_overlap_valid = valid_df.parent_asin_in_train.sum() * 100 // len(valid_df)
    user_and_product_overlap_valid = valid_df[(valid_df.user_in_train == True) & (valid_df.parent_asin_in_train == True)].shape[0] * 100 // len(valid_df)
    user_overlap_test = test_df.user_in_train.sum() * 100 // len(test_df)
    product_overlap_test = test_df.parent_asin_in_train.sum() * 100 // len(test_df)
    user_and_product_overlap_test = test_df[(test_df.user_in_train == True) & (test_df.parent_asin_in_train == True)].shape[0] * 100 // len(test_df)

    rows_list.append([category, no_of_traning_records, no_of_valid_records, no_of_test_records, \
        user_overlap_valid, product_overlap_valid, user_and_product_overlap_valid,\
        user_overlap_test, product_overlap_test, user_and_product_overlap_test])
```

```python
_df = pd.DataFrame(rows_list, columns=cols)
```
